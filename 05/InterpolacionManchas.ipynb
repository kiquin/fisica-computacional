Primero importamos el archivo de datos, seleccionando las columnas que nos interesan: El tiempo de medición y el número de manchas solares para ese tiempo

```python
from scipy import interpolate

data = loadtxt('sparse_sample_monthrg.dat')
year = data[:,0]
y = data[:,2]
month = data[:,1]

x = year + month/12 #Sumo los meses de forma decimal al año

x_new = linspace(x.min(),x.max(), 5000)

f1 = interpolate.interp1d(x, y)
f2 = interpolate.interp1d(x, y, 'cubic')
f3 = interpolate.interp1d(x, y, 'nearest')


y_lineal = f1(x_new) #interpolacion lineal

y_cubic = f2(x_new) #interpolacion con splines cubicos

y_near = f3(x_new) #interpolacion constante
```

```python
from scipy.fftpack import fft, fftfreq, ifft

fft_lineal = fft(y_lineal)

fft_cubic = fft(y_cubic)

fft_near = fft(y_near)

T = (x.max() - x.min())/5000

freq = fftfreq(5000,T)
```

Ahora, podemos graficar los espectros de potencia para cada una de las interpolaciones realizadas

```python
figure(figsize(16,4))

subplot(131)
title('Constante')
semilogy(freq, abs(real(fft_near))**2)

subplot(132)
title('Lineal')
semilogy(freq, abs(real(fft_lineal))**2,'r')

subplot(133)
title('Cubica')
semilogy(freq, abs(real(fft_cubic))**2,'g')
```

Una vez hemos eliminado los datos, procedemos a realizar la transformada inversa de cada señal, con el fin de reconstruir la señal original a partir de los nuevos datos

```python
new_yn = ifft(fft_n)
new_yl = ifft(fft_l)
new_yc = ifft(fft_c)
```

A partir de los datos nuevos, podemos realizar gráficas en donde los comparamos con los datos originales

```python
figure(figsize(16,18))

subplot(311)
title('Interpolacion Constante')
plot(x,y, 'k.', label = 'Datos Originales')
plot(x_new, real(new_yn), 'b', label = 'Reconstruccion')
legend(loc='lower right')

subplot(312)
title('Interpolacion Lineal')
plot(x,y, 'k.', label = 'Datos Originales')
plot(x_new, real(new_yl), 'r', label = 'Reconstruccion')
legend(loc='lower right')

subplot(313)
title('Interpolacion Cubica')
plot(x,y, 'k.', label = 'Datos Originales')
plot(x_new, real(new_yc), 'g', label = 'Reconstruccion')
legend(loc='lower right')
```
