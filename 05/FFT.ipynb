Primero, cargamos el archivo que contiene las 24 señales del encefalograma, obteniendo sus dimensiones, con el fin de crear un vector de tiempo 'x', y un arreglo nuevo del mismo tamaño que la entrada, en el cual guardaremos los datos de la transformada de fourier

```python
data = loadtxt('sampled+ma0844az_1-1+_data.txt')
n = len(data[:,1])
m = len(data[1,:])
x = linspace(1,n,n)
fourier = empty([n,m], dtype=complex)
```

```python
figure(figsize=(16,4)) 
xlabel('Tiempo (Segundos)')
ylabel('Amplitud')
for i in range(0,m):
    plot(x,data[:,i])
```

Importamos las secuencias para realizar la transformada rápida de fourier y su inversa. Usando una función iterativa, realizamos la trasformada de fourier de cada una de las señales, columna por columna, y las guardamos en el nuevo arreglo, llamado 'fourier'. Obtenemos el vector de frecuencias utilizando la secuencia 'fftfreq', para un periodo de muestreo de 1 minuto

```python
from scipy.fftpack import fft, fftfreq, ifft

for i in range(0,m):
    fourier[:,i] = fft(data[:,i])
    
freq = fftfreq(n,1)
```

Una vez tenemos el resultado y el vector de frecuencias, procedemos a graficar los resultados obtenidos como un espectro de potencias, para las frecuencias positivas. Para visualizar los resultados de manera más fácil, realizamos 4 gráficas, cada una con el espectro de potencia de 6 señales, para un total de 24

```python
mitad = ceil(n/2.0)
fourier_mitad = fourier[:mitad,:]
freq_mitad = freq[:mitad]

figure(figsize=(16,10)) #Creo la figura lo suficientemente grande

#Grafica señales 1 a 6

subplot(221)
xlabel('Frecuencia')
ylabel('Potencia')

for i in range(0,m/4):
    loglog(freq_mitad, abs(fourier_mitad[:,i])**2) 
    
#Grafica señales 7 a 12

subplot(222)
xlabel('Frecuencia')
ylabel('Potencia')

for i in range(m/4,m/2):
    loglog(freq_mitad, abs(fourier_mitad[:,i])**2) 
    
    
#Grafica señales 13 a 18

subplot(223)
xlabel('Frecuencia')
ylabel('Potencia')

for i in range(m/2,3*m/4):
    loglog(freq_mitad, abs(fourier_mitad[:,i])**2) 
    
    
#Grafica señales 19 a 24

subplot(224)
xlabel('Frecuencia')
ylabel('Potencia')

for i in range(3*m/4,m):
    loglog(freq_mitad, abs(fourier_mitad[:,i])**2) 
    
suptitle('Espectro de Potencias', fontsize = 24)
```

Se realiza el siguiente método iterativo en donde procesamos columna por columna de las señales en frecuencia de fourier, utilizando la secuencia 'sort' de numpy para obtener un vector con los 10 mayores valores en la columna. Conociendo el décimo más grande (posición 0 del vector 'mayores'), procedemos a eliminar los valores más pequeños, convirtiéndolos a ceros. Obtenemos columnas en donde sólo quedan los 10 valores más grandes.

```python
import copy

fft = copy.deepcopy(fourier) #Copio el vector para evitar conflictos con las graficas anteriores

for i in range(0,m):
    mayores = sort(abs(fft[:,i]), axis=None)[-10:]
    fft[:,i][abs(fft[:,i])<mayores[0]] = 0
```

Finalmente, para realizar la transformada inversa de fourier de nuestras señales nuevas, creamos un nuevo arreglo vacío, en el cual consignamos los datos que resultan del uso de la secuencia 'ifft'.

```python
new = empty([n,m], dtype=complex)
for i in range(0,m):
    new[:,i] = ifft(fft[:,i])
```

De esta manera, hemos reconstruido las señales a partir de sus 10 frecuencias con mayor potencia. Para visualizar los resultados, escogemos 3 señales entre las 24, comparando la original con nuestra reconstrucción

```python
figure(figsize=(18,4)) 

#Señal 6
subplot(131)
plot(x,real(new[:,5]), 'b', label = 'Reconstruida')
plot(x,data[:,5], 'r', label = 'Original')
legend(loc='upper left')

#Señal 13
subplot(132)
plot(x,real(new[:,12]), 'b')
plot(x,data[:,12], 'r')

#Señal 22
subplot(133)
plot(x,real(new[:,21]), 'b')
plot(x,data[:,21], 'r')
```

Para saber cómo se compara la señal reconstruida a la original, calculamos el valor chi cuadrado de la misma con respecto a la original, para cada una de las señales. Creamos un arreglo de 24 ceros, en donde se van sumando las diferencias al cuadrado entre las señales nuevas y originales. Este arreglo contiene los valores de la estadística Chi Cuadrado para cada señal

```python
chi2 = zeros(m)

for i in range(0,m):
    for j in range(0,n):
        chi2[i] += ((data[j,i] - real(new[j,i]))**2)

for i in range(0,m):
    print 'Chi2 Señal ' + str(i) + ' = ' +  str(chi2[i]/n).rstrip('0')
```
