```python
n_points = 200
T = 0.15
c = 1.0

x = linspace(0,2,n_points)
u_inic = 3.0*sin(pi*x)
u_inic[where(x == 1)] = 0

delta_x = x[1] - x[0]
delta_t = 0.0005

beta = c/(delta_x/delta_t)

print beta
```

## Metodo de diferencias finitas (Diferencias centradas)

```python
u_futuro = zeros(n_points)
import copy
u_pasado = copy.deepcopy(u_inic)
u_presente = copy.deepcopy(u_inic)
n_time = int(T/delta_t)

t = linspace(0,T,n_time)
```

```python
grid_x, grid_t = mgrid[0:2:200j, 0:0.15:300j]
grid_y = empty((200,300))

grid_y[:,0] = u_pasado
```

```python
for j in range(n_time):
    for i in range(1,n_points-1):
        u_futuro[i] = u_pasado[i] - beta * (u_presente[i+1]**2 - u_presente[i-1]**2) / 2.0
    u_pasado = copy.deepcopy(u_presente)
    #variable para el valor de la iteración en el presente
    u_presente = copy.deepcopy(u_futuro)
    grid_y[:,j] = copy.deepcopy(u_presente)
```

```python
plot(x,u_presente,linewidth=1.5)
plot(x,u_inic, 'g--',linewidth=3.0)
title('$Diferencias\ Centradas$', size=20)
show
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = figure(figsize=(12,7))
ax = fig.add_subplot(1,1,1, projection='3d')
ax.plot_surface(grid_x,grid_t,grid_y)
title('$Diferencias\ Centradas$', size=20)
xlabel('Espacio', size=16)
ylabel('Tiempo', size=16)
```

## Método de diferencias finitas basado en el esquema de Lax-Wendroff

```python
u_p = zeros(n_points)
u_f = copy.deepcopy(u_inic)
```

```python
g_y = empty((200,300))

g_y[:,0] = copy.deepcopy(u_inic)
```

```python
for s in range(n_time): 
    u_p = copy.deepcopy(u_f)
    for j in range(1,n_points-1):
         u_f[j]= u_p[j]-(beta/4.0)*(u_p[j+1]**2 - u_p[j-1]**2) + ((beta**2)/8.0)*((u_p[j+1] + u_p[j])*(u_p[j+1]**2  -  u_p[j]**2) - ( u_p[j] + u_p[j-1])*( u_p[j]**2 - u_p[j-1]**2 ))
    g_y[:,s] = copy.deepcopy(u_f)
```

```python
plot(x,u_p, linewidth=1.5)
plot(x,u_inic,'g--',linewidth=3.0)
title('$Lax-Wendroff$', size=20)
show
```

```python
fig = plt.figure(figsize=(12,7))
ax = fig.add_subplot(1,1,1, projection='3d')
ax.plot_surface(grid_x,grid_t,g_y)
title('$Lax-Wendroff$', size=20)
xlabel('Espacio', size=16)
ylabel('Tiempo', size=16)
```

## Comentarios

Ambos métodos llegan a una figura similar. Sin embargo, la principal diferencia la podemos evidenciar en las gráficas de la condición final de ambos métodos (en t = 0.15s). En el resultado obtenido a partir del método de diferencias centradas, se puede apreciar la existencia de ruido numérico en las oscilaciones que ocurren cerca de los valores máximo y mínimo de la curva. Por el contrario, al utilizar el método de Lax-Wendroff, la amplitud del ruido es mucho más tenue. 

Esto se debe a que el primer método utiliza una aproximación para de las derivadas con el método de diferencias centradas, preciso hasta primer orden en tiempo y espacio, lo cual genera los errores vistos en las figuras. Mientras que el esquema de Lax-Wendroff, al realizar una expansión de segundo orden de la función, es mucho más preciso y estable, y cómo evidenciamos, tiene mucho menor ruido numérico.
